## Frage 4 

## Welche 15 Topics oder Genre kommen am häufigsten vor (Black History Month Dataset, Topics title)


```python
import urllib.request
import pandas as pd
# Pandas zur Datenanalyse importieren
```

```python
url = "https://raw.githubusercontent.com/MrsMaulwurf/2022-BDK6.5.3-P5_Gruppe_5/main/topics.csv"
# Variable url einführen mit Wert Link zur csv-Datei in GitHub

topics_file = "topics_black_history_month.csv"
# Variable topics_file einführen
```

```python
urllib.request.urlretrieve(url, topics_file)

# library urllib und modul request mit methode urlretrieve (mit den 2 Argumenten für Start und Ziel)
```

```python
topics_file
```

```python
topics = pd.read_csv(topics_file)

# csv einlesen und dataframe erzeugen
```

```python
topics
```

```python
topics.columns
# gibt spaltennamen aus
```

```python
topics["Genre"]
# gibt Spalte "Genre" aus
```

```python
genre_count = topics["Genre"].value_counts()
# identische Einträge (Genre) gruppieren, zählen und nach Häufigkeit anordnen
```

```python
print(genre_count)
```

```python
top15 = genre_count.head(15)
# Ausgabe der 15 häufigsten Genres der Zieldatei
```

```python
print(top15)
```

```python
top15vis = top15.to_frame()
# series in dataframe umwandeln
```

```python
top15vis.plot.barh(color="purple")
# Visualisierung in einem Balkendiagramm
```

```python
top15vis.plot.pie(subplots=True, figsize=(11, 12))

# Umwandlung des Balkendiagramms in ein Kuchendiagramm
```

# Erkenntnisse
- Das häufigste Genre ist "Biography", gefolgt von "History". 

# Probleme und Verbesserungen

- Das dritthäufigste Genre ist "Source", was kein Genre ist
- Die Genres sind nicht komplett bereinigt und dementsprechend sind die Ergebnisse abweichend (z.B. Biography; History)
